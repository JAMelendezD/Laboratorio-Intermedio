```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special, optimize
```

```python
datos = np.genfromtxt('datos.csv', delimiter = ',', skip_header = 1)
```

```python
d1 = datos[:,0]
d1 = d1[np.logical_not(np.isnan(d1))]
m1 = datos[:,1]
m1 = m1[np.logical_not(np.isnan(m1))]
d2 = datos[:,3]
m2 = datos[:,4]
a,b,c = np.polyfit(m1, d1, 2)
md2,bd2 = np.polyfit(m2, d2, 1)
```

```python
plt.scatter(m1, d1, c='k', label=r'$0\mu m$')
plt.scatter(m2, d2, c='b', label=r'$50\mu m$')
plt.plot(m2, m2*md2+bd2, '--c')
plt.plot(m1,a*m1**2+b*m1+c, '--r')
#plt.errorbar(voltaje, d1, yerr=0.00125 ,xerr=0.0012 ,ecolor='k',fmt='ko')
#plt.errorbar(voltaje, d2, yerr=0.00125 ,xerr=0.0012 ,ecolor='b',fmt='o')
plt.text(20,20,r'$y=-0.005x^2+0.65x-3.0$',color='k')
plt.text(50,8,r'$y=0.317x+0.57$',color='b')
#plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
#plt.xlim(0.0,0.019)
#plt.ylim(0.0,0.06)
plt.title(r'$minimos$ vs $\Delta x$')
plt.ylabel(r'$\Delta x (\mu m)$')
plt.xlabel(r'$minimos(conteo)$')
plt.legend(loc =2)
plt.show()
#plt.savefig('longitud.pdf', format='pdf', dpi=1200)
```
