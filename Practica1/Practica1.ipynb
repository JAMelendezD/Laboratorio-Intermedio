```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 154
R = 0.2
k = ((32*np.pi*N)/(5*np.sqrt(5)*R)*10**-7)**2
print(k)
```

```python
datos = np.genfromtxt('datos.csv', delimiter = ',', skip_header = 1)
```

```python
voltaje2cm = datos[:,1]
corriente2cm = datos[:,0]
voltaje3cm = datos[:,3]
corriente3cm = datos[:,2]
voltaje4cm = datos[:,5]
corriente4cm = datos[:,4]
voltaje5cm = datos[:,7]
corriente5cm = datos[:,6]

campoM2cm = np.zeros(len(corriente2cm))
campoM3cm = np.zeros(len(corriente2cm))
campoM4cm = np.zeros(len(corriente2cm))
campoM5cm = np.zeros(len(corriente2cm))

for i in range(len(corriente2cm)):
    campoM2cm[i] = ((corriente2cm[i]*32*np.pi*N)/(5*np.sqrt(5)*R)*10**-7)**2
    campoM3cm[i] = ((corriente3cm[i]*32*np.pi*N)/(5*np.sqrt(5)*R)*10**-7)**2
    campoM4cm[i] = ((corriente4cm[i]*32*np.pi*N)/(5*np.sqrt(5)*R)*10**-7)**2
    campoM5cm[i] = ((corriente5cm[i]*32*np.pi*N)/(5*np.sqrt(5)*R)*10**-7)**2
    
m2cm,b2cm = np.polyfit(campoM2cm, voltaje2cm, 1)
m3cm,b3cm = np.polyfit(campoM3cm, voltaje3cm, 1)
m4cm,b4cm = np.polyfit(campoM4cm, voltaje4cm, 1)
m5cm,b5cm = np.polyfit(campoM5cm, voltaje5cm, 1)
```

```python
plt.scatter(campoM2cm, voltaje2cm, c='k', label=r'$2cm$')
plt.scatter(campoM3cm, voltaje3cm, c='r', label=r'$3cm$')
plt.scatter(campoM4cm, voltaje4cm, c='orange', label=r'$4cm$')
plt.scatter(campoM5cm, voltaje5cm, c = 'b', label=r'$5cm$')
plt.plot(campoM2cm, campoM2cm*m2cm+b2cm, '--c')
plt.plot(campoM3cm, campoM3cm*m3cm+b3cm, '--c')
plt.plot(campoM4cm, campoM4cm*m4cm+b4cm, '--c')
plt.plot(campoM5cm, campoM5cm*m5cm+b5cm, '--c')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.xlim(0.0,0.0000070)
plt.title(r'$Voltaje$ vs $B^2$')
plt.ylabel(r'$Voltaje(V)$')
plt.xlabel(r'$B^2(T^2)$')
plt.legend(loc =4)
plt.show()
plt.savefig('VB.pdf', format='pdf', dpi=1200)
```
